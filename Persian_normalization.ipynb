```python
import pandas as pd
import string
import re

def remove_punctuation_and_normalize(text):
    """
    Remove both English and Persian punctuation marks from text,
    handle diacritics, normalize "می" and "نمی", separate specific suffixes,
    replace specific words, and normalize characters based on dictionary_mapping.
    """
    # Define Persian punctuation marks and special characters
    persian_punctuations = '''`÷×؛<>٬٫٪٭,٫،؟""⟪⟫«»٬٫:/.-!()٭،؟[]⟨⟩~*}٪{×+|️•←→‌'''
    
    # Create translator for English punctuation
    translator = str.maketrans('', '', string.punctuation)
    
    # First, remove English punctuation
    text = text.translate(translator)
    
    # Then remove Persian punctuation and special characters
    for char in persian_punctuations:
        text = text.replace(char, ' ')
    
    # Remove diacritics (harakat)
    text = re.sub(r'[\u064B-\u0652]', '', text)
    
    # Normalize spaces for consistency
    text = ' '.join(text.split())
    


       
    # Dictionary mapping for character normalization
    dictionary_mapping = {
        'ك': 'ک', 'دِ': 'د', 'بِ': 'ب', 'زِ': 'ز', 'ذِ': 'ذ', 'شِ': 'ش', 'سِ': 'س', 'ى': 'ی',
        'ي': 'ی', 'أ': 'ا', 'ؤ': 'و', "ے": "ی", "ۀ": "ه", "ﭘ": "پ", "ﮐ": "ک", "ﯽ": "ی",
        "ﺎ": "ا", "ﺑ": "ب", "ﺘ": "ت", "ﺧ": "خ", "ﺩ": "د", "ﺱ": "س", "ﻀ": "ض", "ﻌ": "ع",
        "ﻟ": "ل", "ﻡ": "م", "ﻢ": "م", "ﻪ": "ه", "ﻮ": "و", 'ﺍ': "ا", 'ة': "ه",
        'ﯾ': "ی", 'ﯿ': "ی", 'ﺒ': "ب", 'ﺖ': "ت", 'ﺪ': "د", 'ﺮ': "ر", 'ﺴ': "س", 'ﺷ': "ش",
        'ﺸ': "ش", 'ﻋ': "ع", 'ﻤ': "م", 'ﻥ': "ن", 'ﻧ': "ن", 'ﻭ': "و", 'ﺭ': "ر", "ﮔ": "گ"
    }
    
    # Apply character replacements from dictionary_mapping
    for old_char, new_char in dictionary_mapping.items():
        text = text.replace(old_char, new_char)
    
    
    # Dictionary of specific replacements
    # ..............
    
    # Apply specific replacements
    for old, new in replacements.items():
        text = re.sub(r'\b' + re.escape(old) + r'\b', new, text)

    
    # Concatenate "می" or "نمی" with the following word
    
    text = re.sub(r'\b(می|نمی)\s+', r'\1', text)
    
    # Separate suffixes "ها", "های", "هایمان" from the root word
    text = re.sub(r'(\S+)(ها|های|هایمان)\b', r'\1 \2', text)
    
    text = re.sub(r'(?<=\S)هه(?=\S)', 'ه ه', text)

    # Separate "ی" from words ending with "هی"
    text = re.sub(r'(\S+ه)ی\b', r'\1 ی', text)
    
    return text

def clean_csv_file(input_path, output_path):
    """
    Read CSV file, clean the 'text' column, and save to new file.
    """
    try:
        # Read CSV file
        df = pd.read_csv(input_path, encoding='utf-8')
        
        # Check if 'text' column exists
        if 'text' not in df.columns:
            raise ValueError("Column 'text' not found in CSV file")
        
        # Apply cleaning function to 'text' column
        df['text'] = df['text'].astype(str).apply(remove_punctuation_and_normalize)
        
        # Save cleaned data to new CSV file
        df.to_csv(output_path, encoding='utf-8', index=False)
        
        print(f"Successfully processed {len(df)} rows")
        print(f"Cleaned file saved to: {output_path}")
        
        # Return sample of cleaned text
        return df['text'].head()
        
    except Exception as e:
        print(f"Error processing file: {str(e)}")
        return None

# Example usage
if __name__ == "__main__":
    input_file = "./cleaned_manifests/filtered_train_file4.csv"
    output_file = "./cleaned_manifests/filtered_train_file5.csv"
    
    # Process the file
    sample_results = clean_csv_file(input_file, output_file)
    
    # Show sample results
    if sample_results is not None:
        print("\nSample of cleaned text:")
        for i, text in enumerate(sample_results, 1):
            print(f"{i}. {text}")
```

```python
#### clean some rows that are more than 200 characters
import csv
import sys

# Increase the field size limit
csv.field_size_limit(sys.maxsize)

# Function to clean up a CSV file
def clean_csv_file(input_file, output_file):
    total_rows = 0
    valid_rows = 0

    with open(input_file, 'r') as file:
        reader = csv.reader(file)
        rows = []
        for row in reader:
            total_rows += 1
            if all(len(str(cell)) <= 200 for cell in row):
                rows.append(row)
                valid_rows += 1
            else:
                print(f"Skipping row {total_rows}: contains field larger than {csv.field_size_limit()} characters")

    with open(output_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)

    print(f"CSV file cleaned up. Wrote {valid_rows} out of {total_rows} rows to {output_file}")

# Example usage
clean_csv_file('./cleaned_manifests/filtered_train_file5.csv', './cleaned_manifests/filtered_train_file6.csv')
```
